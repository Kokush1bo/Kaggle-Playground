### Correlation analysis

`Heatmap`

```python
df['Status'].replace({'D': 1, 'C': 2, 'CL': 3}, inplace=True)
```

```python
numeric_features = df.select_dtypes(include=[np.number])
numeric_features.columns
```

```python
categorical_features = df.select_dtypes(exclude=[np.number])
categorical_features.columns
```

```python
correlation = numeric_features.corr()
print(correlation['Status'].sort_values(ascending=False),'\n')
```

```python
f, ax = plt.subplots(figsize = (12,10 ))
sns.heatmap(correlation, square=True,annot=True)
```

```python
p25 = df['Alk_Phos'].quantile(0.25)
p75 = df['Alk_Phos'].quantile(0.75)
IQR = p75-p25
p25, p75
```

```python
upper_limit = p75 + 3*IQR
lower_limit = p25 - 3*IQR
```

```python
new_df[new_df['Alk_Phos']>upper_limit].count()
# new_df[new_df['Bilirubin']<lower_limit].count()
```

```python
new_df = new_df[new_df['Alk_Phos']<upper_limit]
new_df = new_df[new_df['Alk_Phos']>lower_limit]
new_df.shape
```
